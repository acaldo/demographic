```python
# Filtrar las filas donde 'salary' es '<=50K' y '>50K' respectivamente
lower_df = df[df['salary'] == '<=50K']
higher_df = df[df['salary'] == '>50K']

# Obtener el conteo por 'native-country' para ambas categorías
lower_counts = lower_df['native-country'].value_counts()
higher_counts = higher_df['native-country'].value_counts()

# Combinar los conteos en un DataFrame
combined_df = pd.DataFrame({'count_lower': lower_counts, 'count_higher': higher_counts}).fillna(0)

# Calcular el porcentaje de 'native-country' con '>50K'
combined_df['percent_higher'] = (combined_df['count_higher'] / (combined_df['count_lower'] + combined_df['count_higher'])) * 100

# Obtener el país con el porcentaje más alto
highest_earning_country = combined_df['percent_higher'].idxmax()
highest_earning_country_percentage = combined_df['percent_higher'].max()

print(f"The country with the highest percentage of people earning >50K is {highest_earning_country} with {highest_earning_country_percentage:.1f}%.")
```
